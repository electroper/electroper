## **Indexing and Slicing Using loc Function**


```python
import pandas as pd
```

## Sorting Dataframes

```python
import matplotlib.pyplot as plt
import seaborn as sns
# sets the default style for plotting
sns.set_style("darkgrid")
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

```python
age_sorted = titanic_data.sort_values(by = ["age"])
age_sorted.head()
```

```python
age_sorted = titanic_data.sort_values(by = ["age"], ascending=False)
age_sorted.head()
```

```python
age_sorted = titanic_data.sort_values(by=['age','fare'], ascending = False)
age_sorted.head()
```
